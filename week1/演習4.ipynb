```python
def generate_fizzbuzz(max_num):
  number=[]
  for i in range(1,max_num):
    if i % 3 == 0 and i % 5 == 0:
      i = 'Fizz Buzz'
      number.append(i)
    elif i % 3 == 0:
      i='Fizz'
      number.append(i)
    elif i % 5 == 0:
      i == 'Buzz'
      number.append(i)
    else:
      number.append(i)
  return number
```

```python
generate_fizzbuzz(100)
```
